10 turnajov, na ktor√Ωch bolo odohran√Ωch najviac z√°pasov.

```python
tournaments = results.groupby('tournament').size().reset_index(name='count')

#sort the data by the count in descending order
tournaments = tournaments.sort_values('count', ascending=False)

#select the top 10 values
top_10_tournaments = tournaments.head(10).sort_values('count',ascending=True)

#talking about football so everything will be green
cmap = cm.get_cmap('Greens')

#create a horizontal bar chart
#take log of counts so the lower count tournament arent so faded
plt.barh(top_10_tournaments['tournament'], top_10_tournaments['count'],
            color=cmap(np.log(top_10_tournaments['count']) / np.log(top_10_tournaments['count'].max())))
plt.ylabel('Tournament')
plt.xlabel('Count')
plt.title('Top 10 Tournaments')

# show the chart
plt.show()
```

Medzin√°rodn√© turnaje sa nehraj√∫ ƒçasto, tak≈æe najviac hier je odohran√Ωch ako priateƒæsk√Ω z√°pas.

#### Naj√∫spe≈°nej≈°ie t√≠my (strieƒæanie g√≥lov)

```python
home = results[['home_team', 'home_score']].rename(columns={'home_team':'team', 'home_score':'score'})
away = results[['away_team', 'away_score']].rename(columns={'away_team':'team', 'away_score':'score'})
teams = pd.concat([home, away])

teams_info = teams.groupby('team').agg({'score': ['sum'], 'team': ['size']})
teams_info.columns = ['goals_scored', 'num_matches']

#only take teams with above 5 games to avoid bias
teams_info = teams_info[teams_info['num_matches'] > 5]

teams_info['average'] = teams_info['goals_scored']/teams_info['num_matches']
```

```python
#average goals scored
highest_average = teams_info.sort_values(by='average', ascending=False).head(10)
highest_average.plot(kind='bar', y='average', 
                    color=cmap(highest_average['average']/highest_average['average'].max()),
                    legend=False)
plt.title('Top 10 countries with the highest average goals per game ratio')
plt.xlabel('Country')
plt.ylabel('Average goals per game')
plt.show()
```

```python
#total goals scored
highest_scored = teams_info.sort_values(by='goals_scored', ascending=False).head(10)
highest_scored.plot(kind='bar', y='goals_scored',
                    color=cmap(highest_scored['goals_scored']/highest_scored['goals_scored'].max()),
                    legend=False)
plt.title('Top 10 countries with most goals scored')
plt.xlabel('Country')
plt.ylabel('Total number of goals')
plt.show()
```

Datasetu results prid√°me pr√≠znak `winner`, ktor√Ω bude na≈°ou vysvetlovanou premennou, a hovor√≠m o tom, ƒçi t√≠m hraj√∫ci ako dom√°ci zv√≠≈•azil.

```python
#take draw as a loss
results['winner'] = results.apply(lambda x : 1 if x['home_score'] > x['away_score'] else 0, axis=1)
```

```python
total_matches_home = results.groupby('home_team').size().reset_index(name='total_matches')
total_matches_away = results.groupby('away_team').size().reset_index(name='total_matches')

#create new column for win_count
winner_home = pd.DataFrame(results.groupby('home_team').size() - results[results['winner'] == 0].groupby('home_team').size())
winner_away = pd.DataFrame(results[results['winner'] == 0].groupby('away_team').size())

#merge column into dataframe
total_matches_home = total_matches_home.merge(winner_home.reset_index(),left_on='home_team',right_on='home_team')
total_matches_away = total_matches_away.merge(winner_away.reset_index(),left_on='away_team',right_on='away_team')

total_matches_home.rename(columns={0: 'win_count'}, inplace=True)
total_matches_away.rename(columns={0: 'win_count'}, inplace=True)
```

#### Naj√∫spe≈°nej≈°ie t√≠my (vyhrat√© z√°pasy)
Tu t√≠my rozdel√≠me podƒæa toho, koƒæko z√°pasov vyhr√°vali v dom√°com a vo vonkaj≈°om prostred√≠.
Budeme pri tom sledova≈• t√≠my, ktor√© vyhrali najviac z√°pasov, a t√≠my, ktor√© vyhrali s najvy≈°≈°√≠m % vyhrat√Ωch z√°pasov.

```python
#top 10 teams by wins
win_top_10_home = total_matches_home.sort_values(by='win_count',ascending=False).head(10)

win_top_10_home.plot(x='home_team',y='win_count',kind='bar', 
                    color =cmap(win_top_10_home['win_count']/win_top_10_home['win_count'].max()))
plt.title('Top 10 Countries with most wins at home')
plt.xlabel('Country')
plt.ylabel('Number of wins')
plt.show()
```

```python
#calculate win percentage
total_matches_home['win_percentage'] = total_matches_home['win_count'] / total_matches_home['total_matches'] * 100

#sort by win percentage and select top 10
percent_top_10_home = total_matches_home.sort_values(by='win_percentage',ascending=False).head(10)

#visualize using bar chart
percent_top_10_home.plot(x='home_team',y='win_percentage',kind='bar', 
                    color =cmap(percent_top_10_home['win_percentage']/percent_top_10_home['win_percentage'].max()))
plt.title('Top 10 Countries with highest win percentage at home')
plt.xlabel('Country')
plt.ylabel('Win Percentage')
plt.show()
```

```python
#top 10 teams by wins
win_top_10_away = total_matches_away.sort_values(by='win_count',ascending=False).head(10)

win_top_10_away.plot(x='away_team',y='win_count',kind='bar', 
                    color =cmap(win_top_10_away['win_count']/win_top_10_away['win_count'].max()))
plt.title('Top 10 Countries with most wins away')
plt.xlabel('Country')
plt.ylabel('Number of wins')
plt.show()
```

```python
#avoid bias by countries with just 1 away game
total_matches_away = total_matches_away[total_matches_away['win_count'] > 5]

#calculate win percentage
total_matches_away['win_percentage'] = total_matches_away['win_count'] / total_matches_away['total_matches'] * 100

#sort by win percentage and select top 10
percent_top_10_away = total_matches_away.sort_values(by='win_percentage',ascending=False).head(10)

#visualize using bar chart
percent_top_10_away.plot(x='away_team',y='win_percentage',kind='bar', 
                    color =cmap(percent_top_10_away['win_percentage']/percent_top_10_away['win_percentage'].max()))
plt.title('Top 10 Countries with highest win percentage away')
plt.xlabel('Country')
plt.ylabel('Win Percentage')
plt.show()
```

#### Evol√∫cia teraj≈°√≠ch 10 najlep≈°√≠ch t√≠mov od roku 2018

V ≈æiadnom z na≈°√≠ch datasetoch sa nenach√°dzaj√∫ ch√Ωbaj√∫ce d√°ta, tak≈æe ich nemus√≠me dopƒ∫≈àa≈•.

D√°tumy v datasetoch results a ranking s√∫ r√¥zne, tak≈æe nahrad√≠me d√°tumy v datasete results najbli≈æs√≠m d√°tumom danej krajiny z datasetu ranking. To n√°m veƒæmi zjednodu≈°√≠ ƒèal≈°iu pr√°cu s d√°tami.

```python
def closest_date(date, country):
    """finds closest date from dataset ranking"""

    #only search for date where current country was ranked
    country_ranks = ranking[ranking['country_full'] == country]

    #if country is not in ranks return original date
    if not country_ranks.shape[0]:
        return date

    index = (country_ranks['rank_date'] - date).abs().idxmin()
    return country_ranks.loc[index]['rank_date']


results['home_rank_date'] = results.apply(lambda x: closest_date(x['date'], x['home_team']), axis=1)
results['away_rank_date'] = results.apply(lambda x: closest_date(x['date'], x['away_team']), axis=1)
```

Spojenie datasetov results a ranking do jedn√©ho.

```python
#merge rankings for home teams
df=results.merge(ranking[['country_full', 'total_points', 'rank', 'rank_date']],
                        left_on=['home_rank_date', 'home_team'],
                        right_on=['rank_date', 'country_full']).drop(['rank_date', 'country_full'], axis=1)

#rename columns to not mix them up
df = df.rename(columns={'total_points': 'home_total_points', 'rank': 'home_rank'})


#merge rankings for away teams
df=df.merge(ranking[['country_full', 'total_points', 'rank', 'rank_date']],
                        left_on=['away_rank_date', 'away_team'],
                        right_on=['rank_date', 'country_full']).drop(['rank_date', 'country_full'], axis=1)

df = df.rename(columns={'total_points': 'away_total_points', 'rank': 'away_rank'})

#hove a column date for both home and away
df.drop(['home_rank_date','away_rank_date'],axis=1 , inplace = True)
```

### Feature engineering
D√°tam prid√°me nov√© pr√≠znaky vytvoren√© z t√Ωch p√¥vodn√Ωch. Tieto pr√≠znaky n√°m v√Ωrazne pom√¥≈æu pri tr√©novan√≠ n√°≈°ho modelu.

```python
#difference in ranks of the countries plazing the match
df['rank_dif'] = df['home_rank'] - df['away_rank']
#difference in FIFA points
df['point_dif'] = df['home_total_points'] - df['away_total_points']
df['rank_avg'] = (df['home_rank'] + df['away_rank']) / 2
#1 if match was important (wasnt just a friendly game)
df['is_important'] = df.apply(lambda x : True if x['tournament'] != 'Friendly' else False, axis=1)
df['is_worldcup'] = df.apply(lambda x : True if x['tournament'] == 'FIFA World Cup' else False, axis=1)
```

# Tr√©novanie modelov

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import roc_curve,plot_confusion_matrix
from sklearn.inspection import permutation_importance
from sklearn.metrics import roc_auc_score
import sklearn.metrics as metrics
```

Rozdel√≠me d√°ta na tr√©novaciu a validaƒçn√∫ mno≈æinu. Na≈°ou vysvetƒæovanou premennou bude `winner`.

```python
rd_seed = 42
#drop columns with strings
X_train, X_val, y_train, y_val = train_test_split(df.drop(columns=['winner','home_team','away_team']), 
                                                    df['winner'], 
                                                    test_size=0.25, 
                                                    random_state=rd_seed)
```

Pre ka≈æd√Ω model sa pok√∫sime n√°js≈• ƒço najlep≈°iu kombin√°ciu **hyperparametrov**.

```python
def hyperparameter_tuning(param, Model):
    """function that finds the best combination of hyperparameters for a given model"""

    #create all combinations of given hyperparameters
    param_grid = ParameterGrid(param)

    val_acc = []

    for params in param_grid:
        model = Model(**params)
        model.fit(X_train, y_train)

        #store accuracy using current hyperparameters
        val_acc.append(metrics.accuracy_score(y_val, model.predict(X_val)))
    
    #return the combination with highest accuracy
    return param_grid[np.argmax(val_acc)]
```

Pre ka≈æd√Ω model taktie≈æ zobraz√≠me **roc krivku** a **maticu z√°men**.

```python
def plot_roc_curve(probs):
    """function that draws a roc curve for a given model"""

    fpr, tpr, _ = roc_curve(y_val, probs)

    #visualize the curve
    plt.plot(fpr, tpr, label='ROC', color='limegreen')
    plt.plot([0, 1], [0, 1], linestyle='--', color='darkgreen')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
```

```python
def importance_bar(model):
    """function that draws a bar plot of importance of each feature for a given model"""

    #compute permutation importance
    result = permutation_importance(model, X_val, y_val, n_repeats=10, random_state=0)
    importance = result.importances_mean

    #visualize the feature importance
    plt.bar(X_train.columns, importance, color='green')
    plt.title('Importance of each feature in model')
    plt.xlabel('Feature name')
    plt.ylabel('Permutation importance')
    plt.show()
```

```python
def test_model(model):
    """puts other functions together and also returns accuracy of the model"""

    model.fit(X_train, y_train)
    probs = model.predict_proba(X_val)[:, 1]
    auc = metrics.roc_auc_score(y_val, probs)
    print('Accuracy score for train data: {0:.5f}'.format(metrics.accuracy_score(y_train, model.predict(X_train))))
    print('Accuracy score for validation data: {0:.5f}'.format(metrics.accuracy_score(y_val, model.predict(X_val))))
    print('AUC score for validation data: {0:.5f}'.format(auc))
    #call all other functions
    
    plot_roc_curve(probs)
    plot_confusion_matrix(model, X_val, y_val, cmap=plt.cm.Greens, normalize = 'all')
    plt.title("Confusion Matrix")
    plt.show()
    importance_bar(model)

    return (metrics.accuracy_score(y_train, model.predict(X_train)), metrics.accuracy_score(y_val, model.predict(X_val)),auc)
```

### Decision Tree

```python
param_dt = {
    'max_depth': range(1, 30), 
    'criterion': ['entropy', 'gini', 'log_loss']}

best_params_dt = hyperparameter_tuning(param_dt, DecisionTreeClassifier)
print(best_params_dt)
```

```python
dt = DecisionTreeClassifier(**best_params_dt)
dt_train, dt_val, dt_auc = test_model(dt)
```

```python
kNN = KNeighborsClassifier(**best_params_kNN)
kNN_train, kNN_val, kNN_auc = test_model(kNN)
```

### Random Forest

```python
param_rf = {
    'n_estimators':[1,2,3,4,5,6,7,9,10,16,32],
    'max_depth': range(1,10)}

best_params_rf = hyperparameter_tuning(param_rf, RandomForestClassifier)
print(best_params_rf)
```

```python
rf = RandomForestClassifier(**best_params_rf)
rf_train, rf_val, rf_auc = test_model(rf)
```

# Majstrovstv√° sveta v Qatare 2022

```python
#keep only the teams participating in the world cup
wc_teams_ranking = ranking.loc[(ranking['rank_date'] == ranking['rank_date'].max()) & (ranking['country_full'].isin(wc_teams['Team'].unique()))].copy()
wc_teams_ranking.set_index('country_full', inplace=True)
```

Keƒè≈æe majstrovstv√° sveta sa hraj√∫ v Qatare (na neutr√°lnej p√¥de) nechceme aby t√≠m hraj√∫ci ako "dom√°ci" mal v√Ωhodu tak ako pravdepodobnos≈• v√Ωhry budeme bra≈• priemer pravdepodobnost√≠, ≈æe vyhr√° t√≠m1 keƒè je oznaƒçen√Ω ako dom√°ci a ≈æe vyhr√° t√≠m1 keƒè je t√≠m2 oznaƒçen√Ω ako dom√°ci.

```python
def predict_match(country1, country2):
    """calculates the chance of country1 winning the match"""

    point_dif = wc_teams_ranking.loc[country1, 'total_points'] - wc_teams_ranking.loc[country2, 'total_points']
    rank_dif = wc_teams_ranking.loc[country1, 'rank'] - wc_teams_ranking.loc[country2, 'rank']
    rank_avg = (wc_teams_ranking.loc[country1, 'rank'] + wc_teams_ranking.loc[country2, 'rank']) / 2

    #create 1 row dataframes for the model
    #all matches are in world cup and important and on neutral field
    country1_home = pd.DataFrame({'rank_dif' : rank_dif, 'point_dif' : point_dif , 'rank_avg' : rank_avg, 'neutral' : 1}, index=[0])
    country2_home = pd.DataFrame({ 'rank_dif' : -rank_dif, 'point_dif' : -point_dif ,'rank_avg' : rank_avg, 'neutral' : 1}, index=[0])

    #predict the outcomes
    country1_home_prob = lr.predict_proba(country1_home)
    country2_home_prob = lr.predict_proba(country2_home)

    #return the average win percentage
    return (country1_home_prob[0][1] + country2_home_prob[0][0]) / 2
```

### Predpovedanie skupinovej f√°zy ≈°ampion√°tu

```python
#create a win probability column for the schedule dataset
schedule['win_prob'] = 0

for index,row in schedule.iloc[:48].iterrows():
    #add win probabilty for each match
    schedule.loc[index, 'win_prob'] = predict_match(row['country1'], row['country2'])
```

```python
#create a dataframe of group stage
group_stage = schedule.merge(wc_teams[['Group','Team']], left_on='country1', right_on='Team')
group_stage.drop('Team', axis=1, inplace=True)
```

```python
for group_letter in sorted(group_stage['Group'].unique()):
    #matches in current group
    group_matches = group_stage[group_stage['Group'] == group_letter]

    print(f'-------Group {group_letter}-------')

    for index, row in group_matches.iterrows():
        if row['win_prob'] > 0.55:
            #country1 won
            print(f'{row["country1"]} vs {row["country2"]} : {row["country1"]} wins with probability of {round(row["win_prob"], 2)}')
        elif row['win_prob'] < 0.45:
            #country2 won
            print(f'{row["country1"]} vs {row["country2"]} : {row["country2"]} wins with probability of {round(1-row["win_prob"], 2)}')
        else:
            print(f'{row["country1"]} vs {row["country2"]} : Draw')
```

```python
#create column for points gained by a team
wc_teams_ranking['points_gained'] = 0

for index,match in group_stage.iterrows():
    #countries get 3 points for a win and 1 point for a draw
    if match['win_prob'] > 0.55:
        wc_teams_ranking.loc[match['country1'],'points_gained'] += 3
    elif match['win_prob'] < 0.45:
        wc_teams_ranking.loc[match['country2'],'points_gained'] += 3
    else:
        wc_teams_ranking.loc[match['country1'],'points_gained'] += 1
        wc_teams_ranking.loc[match['country2'],'points_gained'] += 1
```

```python
#create a knockout stage dataframe
knockout_stage = schedule.iloc[48:].copy()
knockout_stage.reset_index(drop=True,inplace=True)
```

```python
def compare_points(country1, country2):
    """compares place of two countries in their group"""

    if wc_teams_ranking.loc[country1, 'points_gained'] > wc_teams_ranking.loc[country2, 'points_gained']:
        return -1
    elif wc_teams_ranking.loc[country1, 'points_gained'] < wc_teams_ranking.loc[country2, 'points_gained']:
        return 1
    else:
        #secondary sorting if countries have the same amount of points
        match_prediction = predict_match(country1, country2)
        if match_prediction >= 0.5:
            return -1
        else:
            return 1
```

```python
def insert_group_winner(place, team):
    """inserts a country into the knockout stage dataframe"""

    #check which column the country belongs to
    if knockout_stage.loc[knockout_stage['country1'] == place].shape[0]:
        knockout_stage.loc[knockout_stage['country1'] == place, 'country1'] = team
    else:
        knockout_stage.loc[knockout_stage['country2'] == place, 'country2'] = team
```

Z ka≈ædej skupiny postupuj√∫ prv√© dve mu≈æstv√° do vyraƒèovacej f√°zy turnaja.

```python
from functools import cmp_to_key

for group_letter in sorted(group_stage.Group.unique()):

    #current group
    group_matches = group_stage[group_stage['Group'] == group_letter]
    #all teams in the group
    group_teams = pd.concat([group_matches['country1'], group_matches['country2']]).unique()
    
    #sort teams by thier points
    group_teams = sorted(group_teams, key=cmp_to_key(compare_points))

    #first place
    insert_group_winner('1' + group_letter.upper(), group_teams[0])
    #second place
    insert_group_winner('2' + group_letter.upper(), group_teams[1])

    print(f'-------Group {group_letter}-------')
    for team in group_teams:
        print(team,':', wc_teams_ranking.loc[team, 'points_gained'], 'points')
```

Do vyraƒèovacej f√°zy turnaja postupuje teda 16 t√≠mov, z toho 10 t√≠mov, ktor√© post√∫pili podƒæa n√°≈°ho modelu, post√∫pili aj v skutoƒçnosti.

### Predpovedanie vyraƒèovacej f√°zy ≈°ampion√°tu

```python
def add_win_prob(fr,to):
    """adds win probability to the knockout_stage dataframe"""

    for index,row in knockout_stage.iloc[fr:to].iterrows():
        knockout_stage.loc[index, 'win_prob'] = round(predict_match(row['country1'], row['country2']), 2)
```

```python
def print_results(fr,to,name):
    """prints results of current stage"""

    print(f'-------{name}-------')
    
    for index, row in knockout_stage.iloc[fr:to].iterrows():
        #dont take draws in the knockout stages
        if row['win_prob'] > 0.50:
            print(f'{row["country1"]} vs {row["country2"]} : {row["country1"]} wins with probability of {round(row["win_prob"], 2)}')
            insert_group_winner('W' + str(index+49), row['country1'])
            
            #losers play for 3rd place
            if name=="Semi Finals":
                insert_group_winner('L' + str(index + 49), row['country2'])
        else:
            print(f'{row["country1"]} vs {row["country2"]} : {row["country2"]} wins with probability of {round(1-row["win_prob"], 2)}')
            insert_group_winner('W' + str(index + 49), row['country2'])

            #3rd place
            if name=="Semi Finals":
                insert_group_winner('L' + str(index + 49), row['country1'])
```

```python
add_win_prob(0,8)
print_results(0,8,'Round of 16')

add_win_prob(8,12)
print_results(8,12,'Quarter Finals') 

add_win_prob(12,14)
print_results(12,14,'Semi Finals')

add_win_prob(14,16)
print_results(14,15,'Third Place')

print_results(15,16,'Final')
```

Braz√≠lia turnaj nevyhrala, ale bola oznaƒçovan√° za najv√§ƒç≈°ieho favorita, tak≈æe model dosiahol predpokladan√Ωch v√Ωsledkov.

```python
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout
import pydot
import matplotlib.patches as mpatches
```

```python
#list of labels for knockout stage graph
labels = []

for index,row in knockout_stage.iterrows():
    #third place is not in the ladder
    if row['phase']!='third place':
        
        #string for both countries and their win probability
        label = f"{row['country1']} ({round(row['win_prob'], 2)}) \n {row['country2']} ({round(1-row['win_prob'], 2)})"
        labels.append(label)
```

```python
#balanced tree of depth 3 and every node branches into 2
G = nx.balanced_tree(2, 3)

# layout for the nodes
pos = graphviz_layout(G,prog='twopi')
```

```python
#dictionary of labels for the graph
labels_dict = {}

for index, label in enumerate(list(reversed(labels))):
    #items in dictionary are index : label
    labels_dict[index] = label
```

```python
#each index is a number 1 to 4 based on phase of the tournament
phases = [4 - int(np.log2(i)) for i in range(1, 16)]

#every phase has its own green color
colors = [(0, i/4, 0) for i in phases]

legend_colors = [(0, i/4, 0) for i in (4, 3, 2, 1)]
names = ['Final','Semi Finals','Quarter Finals','Round of 16']
```

```python
#draw the graph
nx.draw(G, pos=pos, with_labels=False, node_size = 5000, node_color = colors, width=8)
nx.draw_networkx_labels(G, pos=pos, bbox=dict(boxstyle="round,pad=0.1", fc="white", ec="black", lw=.5), labels=labels_dict, font_size=12)

plt.title('Knockout stage')
patches = [mpatches.Patch(color=colors[i], label=names[i]) for i in range(4)]
plt.legend(handles=patches)
plt.show()
```
