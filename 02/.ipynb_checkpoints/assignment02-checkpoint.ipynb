```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx

from datetime import datetime, time
```

```python
df = pd.read_csv('d.csv')
df.head()
```

Vytvorime si stlpec duration, ktory bude symbolizovat dlzku cesty v sekundach

```python
def duration(fncDf): 
    cnt = 0
    cnt += (fncDf['arrive_to'].str.split(':',expand = True)[0].astype(int) 
                        - fncDf['depart_from'].str.split(':',expand = True)[0].astype(int))*3600 
    cnt += (fncDf['arrive_to'].str.split(':',expand = True)[1].astype(int) 
                        - fncDf['depart_from'].str.split(':',expand = True)[1].astype(int))*60   
    cnt += fncDf['arrive_to'].str.split(':',expand = True)[2].astype(int) - fncDf['depart_from'].str.split(':',expand = True)[2].astype(int)
    return cnt
df['duration'] = duration(df) 
```

Zmodulime si stlpce depart_from a arrive_to nech nemame cas vacsi ako 24h

```python
def modulo_hours(time_str):
    hours, minutes, seconds = time_str.split(':')
    hours = int(hours) % 24
    return f'{hours:02d}:{minutes}:{seconds}'

df['depart_from'] = df['depart_from'].apply(modulo_hours)
df['arrive_to'] = df['arrive_to'].apply(modulo_hours)
```

```python
def to_time(time_str):
    time = datetime.strptime(time_str, '%H:%M:%S')
    return time.time()

df['depart_from'] = df['depart_from'].apply(to_time)
df['arrive_to'] = df['arrive_to'].apply(to_time)
```

Vyberame si 'route_type' = 0 co su stanice tramvaji a 'is_night' = 1 co su nocne linky

```python
dfNightTram = df.loc[(df['route_type'] == 0) & (df['is_night'] == 1) ]
```

Vytvorime si NetworkX graf kde node su nazvy zastavok a hrana grafu je duration

```python
G = nx.from_pandas_edgelist(dfNightTram, source='stop_from_name', target='stop_to_name', edge_attr='duration', create_using=nx.Graph())
```

## Analyza pomocou mier centrality

```python
import networkx as nx

degrees = nx.degree_centrality(G)

between = nx.betweenness_centrality(G)

closen = nx.closeness_centrality(G)
```

Zobrazime si 5 najlepsich a 5 najhorsich vrcholov podla rozdielnych mier

```python
def fiveNodes(centralities,boolean ,text):
    sorted_centr = sorted(centralities, key=centralities.get, reverse=boolean)
    top_5_degrees = [centralities[node] for node in sorted_centr[:5]]

    plt.barh(range(5), top_5_degrees)
    plt.barh(range(5), top_5_degrees)

    plt.yticks(range(5), sorted_centr[:5])
    plt.xlabel('Node')
    plt.ylabel('Degree')
    plt.title(f'{text[0]} 5 Nodes by {text[1]}')

    plt.show()
```

```python
def find_max_min_mean(dictionary):
    values = dictionary.values()
    max_value = max(values)
    min_value = min(values)
    mean_value = sum(values) / len(values)
    
    max_key = None
    min_key = None
    for key, value in dictionary.items():
        if value == max_value:
            max_key = key
        if value == min_value:
            min_key = key
    
    print(f"Najvyssia hodnota je {round(max_value,6)} na zastavke {max_key}")
    print(f"Najnizsia hodnota je {round(min_value,6)} na zastavke {min_key}")
    print(f"Priemer je {round(mean_value,6)}")
```

```python
fiveNodes(degrees,False,["Top","Degree"])
fiveNodes(degrees,True,["Worst","Degree"])
find_max_min_mean(degrees)
```

##### Stup≈àov√° centralita meria poƒçet spojen√≠ uzla s ostatn√Ωmi uzlami v sieti.  Uzly s vysokou stup≈àovou centralitou s√∫ dobre spojen√© a maj√∫ veƒæa spojen√≠ s ostatn√Ωmi uzlami.

```python
fiveNodes(between,True,["Worst","Betweenness"])
fiveNodes(between,False,["Top","Betweenness"])
find_max_min_mean(between)
```

##### Centralita 'betweenness' meria ked sa vrchol nach√°dza na najkrat≈°ej ceste medzi dvoma vrcholmi. Uzly s vysokou centralitou s√∫ tak√©, ktor√© s√∫ "medzi" mnoh√Ωmi in√Ωmi uzlami v sieti. Naopak centralita 'betweenness' je nulova alebo nizka ked sa nenachadza na najkratsej ceste.

```python
fiveNodes(closen,False,["Top","Closeness"])
fiveNodes(closen,True,["Worst","Closeness"])
find_max_min_mean(closen)
```

##### Centralita bl√≠zkosti meria priemern√∫ vzdialenos≈• uzla od v≈°etk√Ωch ostatn√Ωch uzlov v sieti. Uzly s vysokou centralitou bl√≠zkosti s√∫ tak√©, ktor√© s√∫ umiestnen√© bl√≠zko mnoh√Ωch in√Ωch uzlov v sieti a maj√∫ vysok√© mo≈ænosti ovplyv≈àova≈• ostatn√© uzly.

```python
for node in list(G_week.nodes()):
    if node not in G_weekend:
        G_week.remove_node(node)
for node in list(G_weekend.nodes()):
    if node not in G_week:
        G_weekend.remove_node(node)
```

```python
degreesWeek = nx.degree_centrality(G_week)
betweenWeek = nx.betweenness_centrality(G_week)
closenWeek = nx.closeness_centrality(G_week)
degreesWeekend = nx.degree_centrality(G_weekend)
betweenWeekend = nx.betweenness_centrality(G_weekend)
closenWeekend = nx.closeness_centrality(G_weekend)
```

```python
def get_max_min_mean(dictionary):
    values = dictionary.values()
    max_value = max(values)
    min_value = min(values)
    mean_value = sum(values) / len(values)

    return max_value,min_value,mean_value
```

Najdeme si rozdiel pre vsetky miery centrality

```python
degree_diff = {node: degreesWeekend[node] - degreesWeek[node] for node in G_week.nodes()}
betweenness_diff = {node: betweenWeekend[node] - betweenWeek[node] for node in G_week.nodes()}
closeness_diff = {node: closenWeekend[node] - closenWeek[node] for node in G_week.nodes()}

degree_max, degree_min, degree_mean = get_max_min_mean(degree_diff)
betweenness_max, betweenness_min, betweenness_mean = get_max_min_mean(betweenness_diff)
closeness_max, closeness_min, closeness_mean = get_max_min_mean(closeness_diff)
print("Degree centrality difference:")
print("Max:", round(abs(degree_max),5))
print("Min:", round(abs(degree_min),5))
print("Mean:",round(abs(degree_mean),5))

print("Betweenness centrality difference:")
print("Max:", round(abs(betweenness_max),5))
print("Min:", round(abs(betweenness_min),5))
print("Mean:", round(abs(betweenness_mean),5))

print("Closeness centrality difference:")
print("Max:", round(abs(closeness_max),5))
print("Min:", round(abs(closeness_min),5))
print("Mean:", round(abs(closeness_mean),5))
```
